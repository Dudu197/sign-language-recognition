```python
import pandas as pd
```

```python
base_path = "../00_datasets/dataset_output/include50/"
train_file = "include50_train.txt"
validate_file = "include50_val.txt"
test_file = "include50_test.txt"
```

```python
def read_file(filename):
    with open(filename) as file:
        lines = [line.rstrip() for line in file]
    return lines
```

```python
train_videos = read_file(base_path + train_file)
validate_videos = read_file(base_path + validate_file)
test_videos = read_file(base_path + test_file)
```

```python
def get_category(file):
    return "-".join(file.split("/")[:-1])
```

```python
train_categories = [get_category(i) for i in train_videos]
validate_categories = [get_category(i) for i in validate_videos]
test_categories = [get_category(i) for i in test_videos]
```

```python
train_categories
```

```python
df_train = pd.DataFrame.from_dict({"category": c, "dataset": "train"} for c in train_categories)
df_validate = pd.DataFrame.from_dict({"category": c, "dataset": "validate"} for c in validate_categories)
df_test = pd.DataFrame.from_dict({"category": c, "dataset": "test"} for c in test_categories)
```

```python
df = pd.concat([df_train, df_validate, df_test])
```

```python
df
```

```python
df["dataset"].value_counts()
```

```python
len(df["category"].unique())
```

```python
counts = {}
for category in df["category"].unique():
    df_category = df[df["category"] == category]
    counts[category] = {
        "train": len(df_category[df_category["dataset"] == "train"]),
        "validate": len(df_category[df_category["dataset"] == "validate"]),
        "test": len(df_category[df_category["dataset"] == "test"]),
    }
```

```python
counts
```

```python
df_counts = pd.DataFrame.from_dict(counts).T
```

```python
df_counts
```

```python
df_counts.apply(lambda i: i/i.sum() * 100, axis=1)
```
