```python
import os
import shutil
import pandas as pd
```

```python
dataset_path = "../dataset_output/libras_minds/libras_minds_openpose_80_frames_augmented.csv"
input_path = "../dataset_output/libras_minds/imgs_raw/"
output_path = "../dataset_output/libras_minds/imgs_lopo/"
```

```python
train = pd.read_csv(dataset_path)
train = train.drop("Unnamed: 0", axis=1)
```

```python
df = train[train["frame"] == 1]
```

```python
df["person"] = df["video_name"].apply(lambda i: i.split("_")[1])
```

```python
people = df["person"].unique()
validate_people = people[-1:]
test_people = people[-2:-1]
train_people = people[:-2]
```

```python
def create_folder(path):
    if not os.path.exists(path):
        os.mkdir(path)
```

```python
def copy_file(video_name, category, destination):
    src = os.path.join(input_path, category, video_name + ".png")
    dst = os.path.join(output_path, destination, category, video_name + ".png")
    create_folder(os.path.join(output_path, destination, category))
    shutil.copyfile(src, dst)
```

```python
df_train = df[df["person"].isin(train_people)]
```

```python
df_test = df[(df["person"].isin(test_people) & (~df["video_name"].str.startswith("augmented")))]
df_validate = df[(df["person"].isin(validate_people) & (~df["video_name"].str.startswith("augmented")))]
```

```python
def copy_df(df_copy, destination):
    for row in df_copy.iterrows():
        video_name = row[1]["video_name"]
        category = str(row[1]["category"])
        copy_file(video_name, category, destination)
```

```python
copy_df(df_train, "train")
copy_df(df_test, "test")
copy_df(df_validate, "validate")
```
